Now, we're going to get into if- else statements. this is the really where programming becomes cool, because at the heart of al programming, you have to be able to tell your program "if we've got one situation, do one thing, if we've got another situation do another thing". For example, if you want to make a videogame, you might have a conditional that is something like "if the enemy's health is above 0, do nothing, but once the enemy's health is below zero, run whatever piece of code you need to kill that enemy"

Actually, let's try that one out below: we're going to imagine we've got an enemy called a "zergoid". We want to write a little piece of code that keeps track of his health, and if it drops below zero, we print "the zergoid is dead"

we're gonna start with the zergoid having 20 health

you're going to notice that between "zergoid and health" is a "_" symbol. It's called an underscore, and if you're creating a variable name and you want spaces, you need to put an uncerscore instead or else python doesn't like it 



```python
zergoid_health= 20

if zergoid_health > 0 :
    print("the zergoid is still alive")
else: 
    print("the zergoid is dead")
```

so, let's break that down. every conditional statement starts with "if". After if is what we call a conditional. looking at just "zergoid_health > 0", the way to understand this is that we are asking the computer to check whether the variable "zergoid health" is greater than zero. if it is, the computer will go and run the code that is under the "if statement"

if that initial statement is false, it will run the statement under the else

OK, let's try something else: what if we want more than two conditions. that seems like something that's important, right? Like, sometimes in life you need to give a more complex set of instructions. Well, we can totally do that. 

what we are looking for is the "elif" statement. It goes after the if statement and before the else statement. you can put another 

let's try another example-- sometimes, right before your enemy dies, it gets supercharged, to give the player an extra challenge. So, let's say that if the zergoids health is below 10 but above 0, it goes into "superdrive mode"



```python
zergoid_health= 8

if zergoid_health >= 10 :
    print("the zergoid is still alive")
elif zergoid_health < 10 and zergoid_health >0:
    print("the zergoid is supercharged")
else: 
    print("the zergoid is dead")
```

so, we've got a few things here. first, ">=" means "greater than or equal to". 

Below, find all the other ways that we can compare values below 



Also, in the "elif" we see an "and". What that means is that computer is going to check whether zergoid_health<10 is true and also whether zergoid-health >0 is true, and only if both of them are true, will it run the code that prints "the zergoid is supercharged

let's try a bit on "and" and "or"

when we put an "and" between two True/ false statements such as:

if a>5 and b>6:
    do something
    

we are essentially saying, "only if both of these are true, do run that code below

however, if we use an "or" statement like below 

if a>5 or b>6:
    do something
    
it means that the computer will accept just a or just b or both, and will still run the code.
